````r
library(ggplot2)
````

````r
athletes <- read.csv("https://rawgit.com/flother/rio2016/master/athletes.csv")
````

````r
summary(athletes)
````

````r
ggplot(data=athletes) + geom_point(mapping=aes(x=height,y=weight))
````

````r
ggplot(athletes) + geom_point(aes(gold,weight))
````

A graphing template in ggplot (in its simplest form) looks like this:
```
ggplot(data = <DATA>) + 
  <GEOM_FUNCTION>(mapping = aes(<MAPPINGS>))
``` 
You can read the previous usage example of this template like this: "Hey ggplot. I want to make a plot out of this dataset called `athletes` *and* I want there to be a layer of `geom_point` (e.g. points) such that variable `height` maps to x-coordinate and `weight` maps to y-coordinate."

It's possible to use more aesthetic mappings:

````r
ggplot(athletes) + geom_point(aes(height, weight, col=sport))
````

````r
ggplot(athletes) + geom_point(aes(height,weight,size=gold))
````

Facets are a way to create a graph type called *small multiples*: same plot but for different sections of the data, according to a categorical variable:

````r
ggplot(athletes) + geom_point(aes(height, weight)) + facet_wrap(~sport)
````

There are other geometric objects besides points:

````r
ggplot(athletes) + geom_smooth(aes(height,weight,col=sport))
````

Different geometries can use the mappings quite differently. But the idea of **mapping variables to aesthetics** stays:

````r
ggplot(athletes) + geom_boxplot(aes(sport, weight))
````

````r
ggplot(athletes) + geom_boxplot(aes(sport, weight)) + coord_flip()
````

One common plot type requires an implicit transformation of the data: the heights of the barplot are not present in the original data! Ggplot will do that automatically for you when you use `geom_bar`. If there is a variable giving the heights directly, use `geom_col` instead.

````r
ggplot(athletes) + geom_bar(aes(sport))
````

````r
ggplot(athletes) + geom_point(aes(height,weight,color=gold))
ggplot(athletes) + geom_point(aes(height,weight,color=gold)) + scale_colour_gradient(low="gray",high="red")
````

````r
ggplot(athletes) + geom_point(aes(height,weight,color=sex))
ggplot(athletes) + geom_point(aes(height,weight,color=sex)) + scale_colour_brewer(palette="Accent")
````

The gray background with the white grid lines is just a default! The reasons for choosing that as the default are presumably good, but it doesn't mean you shouldn't consider something else, particularly when preparing a visualisation for someone else to see:

````r
ggplot(athletes) + geom_point(aes(height,weight,color=sex)) + scale_colour_brewer(palette="Accent") +
theme_void()
````
